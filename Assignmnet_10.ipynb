### What is the distribution  of preferred foot?

```python
sns.countplot(data=df, x='Preferred Foot')


ax = plt.gca()
# Add text labels above the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

### How can we visualize the age distribution of players?

```python
plt.figure(figsize=(12,10))
sns.countplot(df['Age'])

ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

### What is the international reputation of left and right foot?

```python
df.groupby('Preferred Foot')['International Reputation'].value_counts()
```

```python
sns.countplot(x='International Reputation',hue='Preferred Foot',data=df,palette='hls')

ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

### What is the different work rate of the Players?

```python
plt.figure(figsize=(10,12))
sns.countplot(x='Work Rate',data=df)
plt.tight_layout()
df['Work Rate'].value_counts()
```

### Top 10 countries with most number of players

```python
plt.figure(figsize=(10,5))
top10_countries = df['Nationality'].value_counts().head(10)
sns.barplot(x=top10_countries.index,y=top10_countries.values,data=df)

ax = plt.gca()
# Add text labels above the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')
plt.show()
```

### What is the work rate of players segregated by preferred foot?

```python
plt.figure(figsize=(14,7))
sns.countplot(x='Work Rate',hue='Preferred Foot',data=df)

ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

### Number of players at different positions? 



```python
plt.figure(figsize=(14,7))
sns.countplot(x='Position',data=df)


ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

### Age distribution in top countries

```python
df_country_age = df[df['Nationality'].isin(top10_countries.index)]
plt.figure(figsize = (12,6))
sns.boxplot(x = 'Nationality' , y = 'Age' , data = df_country_age, palette='Set2',order=top10_countries.index)


plt.show()
```

### What is the distribution of overall score of players from top countries?

```python
df_country_age = df[df['Nationality'].isin(top10_countries.index)]
plt.figure(figsize = (12,6))
sns.barplot(x = 'Nationality' , y = 'Overall' , data = df_country_age, palette='hls',order=top10_countries.index)


plt.show()
```
