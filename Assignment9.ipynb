### 1. Import packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(color_codes=True)
import warnings
warnings.filterwarnings("ignore")
```

### 2.Load the Dataset

```python
df=pd.read_csv('titanic-training-data.csv')
df.sample(5)
```

### 4.Show the distribution of survived and not survived

```python
sns.countplot(x='Survived',data=df,palette='Set2')
```

 ### 5.Show the distribution across pclass

```python
sns.countplot(x='Pclass',hue='Survived',data=df,palette='hls')

# Calculate the counts for each category
# counts = df.groupby(['Pclass', 'Survived']).size().reset_index(name='counts')

ax = plt.gca()
# Add text labels above the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

 ### 6. Show the distribution of Embarked

```python
sns.countplot(x='Embarked',hue="Survived",data=df,palette='hls')

ax = plt.gca()
# Add text labels above the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

### 7. Show the distribution of Age

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df,x='Age',hue='Survived',palette='Set2')

ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

### 8. Check whether there are outliers in Age
